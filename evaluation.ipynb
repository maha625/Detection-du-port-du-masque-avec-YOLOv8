```python
from ultralytics import YOLO
import os

# 1. Charger le modèle entraîné (best.pt)
# Assurez-vous que le chemin correspond à votre dossier d'entraînement
model_path = 'runs\\detect\\train2\\weights\\best.pt'

if not os.path.exists(model_path):
    print(f"Erreur : Le fichier {model_path} est introuvable.")
    print("Vérifiez que l'entraînement est terminé avec succès.")
else:
    model = YOLO(model_path)

    # 2. Lancer l'évaluation sur le jeu de validation (split défini dans data.yaml)
    print("Lancement de l'évaluation sur les données de validation...")
    metrics = model.val()

    # 3. Affichage des métriques principales
    print("\n--- Résultats de l'Évaluation ---")
    print(f"Précision moyenne (mAP@50) : {metrics.box.map50:.4f}")
    print(f"Précision moyenne (mAP@50-95) : {metrics.box.map:.4f}")
    print(f"Précision (Precision) : {metrics.box.mp:.4f}")
    print(f"Rappel (Recall) : {metrics.box.mr:.4f}")

    print("\nNote : Les graphiques détaillés (matrice de confusion, courbes F1, etc.)")
    print(f"ont été sauvegardés dans : {metrics.save_dir}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 1. Chemin vers le fichier de résultats généré par YOLO
# Par défaut, YOLO crée un fichier 'results.csv' dans le dossier de l'entraînement
results_path = 'runs/detect/train2/results.csv'

if not os.path.exists(results_path):
    print(f"Erreur : Le fichier {results_path} est introuvable.")
    print("Assurez-vous d'avoir lancé l'entraînement (train.py) au moins une fois.")
else:
    # 2. Chargement des données avec Pandas
    df = pd.read_csv(results_path)
    # Nettoyage des noms de colonnes (YOLO ajoute souvent des espaces)
    df.columns = [c.strip() for c in df.columns]

    # 3. Création des graphiques
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle('Métriques de performance du modèle de détection de masques', fontsize=16)

    # Courbe de perte (Loss)
    axs[0, 0].plot(df['epoch'], df['train/box_loss'], label='Train Box Loss')
    axs[0, 0].plot(df['epoch'], df['val/box_loss'], label='Val Box Loss')
    axs[0, 0].set_title('Perte de localisation (Box Loss)')
    axs[0, 0].legend()

    # Courbe de Précision
    axs[0, 1].plot(df['epoch'], df['metrics/precision(B)'], color='green')
    axs[0, 1].set_title('Précision (Precision)')

    # Courbe de Rappel (Recall)
    axs[1, 0].plot(df['epoch'], df['metrics/recall(B)'], color='orange')
    axs[1, 0].set_title('Rappel (Recall)')

    # Courbe mAP@50
    axs[1, 1].plot(df['epoch'], df['metrics/mAP50(B)'], color='red')
    axs[1, 1].set_title('mAP @ 50')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    # 4. Sauvegarde et affichage
    plt.savefig('evaluation_curves.png')
    print("Les courbes personnalisées ont été sauvegardées sous 'evaluation_curves.png'.")
    plt.show()
```

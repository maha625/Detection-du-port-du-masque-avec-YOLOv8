```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# 1. Charger VOTRE modèle (le meilleur issu de l'entraînement)
# Note : Vérifiez bien le chemin, il dépend du nom donné dans train.py
model = YOLO('runs/detect/train2/weights/best.pt')
```

```python
# 2. Exécuter la détection sur une image
# Vous pouvez mettre le chemin d'une photo de vous ou une image du dossier val
image_path = 'test\\telechargement2.jpg' 
results = model(image_path, conf=0.5) # conf=0.5 signifie 50% de certitude minimum
```

```python
# 3. Visualiser le résultat
for r in results:
    img_bgr = r.plot() # Dessine les boîtes et les labels
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,10))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title("Résultat de la détection")
    plt.show()
```

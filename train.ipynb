```python
from ultralytics import YOLO

# 1. Charger un modèle YOLOv8 pré-entraîné (le plus petit et rapide)
model = YOLO('yolov8n.pt')

if __name__ == '__main__':
    # 2. Lancer l'entraînement
    # data : chemin vers votre fichier yaml
    # epochs : nombre de fois que le modèle voit le dataset (25 est bien pour commencer)
    # imgsz : taille des images (640 est le standard)
    # device : 'cpu' ou 0 (si vous avez un GPU NVIDIA)
    results = model.train(
        data='data.yaml', 
        epochs=25, 
        imgsz=640, 
        device='cpu' # Changez par device=0 si vous avez un GPU
    )
    
    print("Entraînement terminé !")
    print("Le meilleur modèle est sauvegardé dans : runs/detect/train/weights/best.pt")
```
